# Pre-procesamiento de Datos
La frecuencia de muestreo fue obtenida de [Ninapro Database](https://ninapro.hevs.ch/instructions/DB6.html), mientras que los demás valores se adquirieron de distintas investigaciones.

```python
# Obtenemos la dimensión del dataframe con los datos de las señales
smeg_dim = smeg_data.shape

'''
fs: Frecuencia de muestreo en Hz

lowcut: Para eliminar frecuencias inferiores al valor espeificado

highcut: Para eliminar frecuencias superiores al valor espeificado

notch_freq: Frecuencia del filtro notch (50 Hz debido a la frecuencia eléctrica)

num_std: Número de desviaciones estándar para reemplazar los valores atípicos,
es decir los límites de corte tanto superior como inferior para las señales EMG
'''
fs = 2000
lowcut = 15
highcut = 500
notch_freq = 50
num_std = 6

functions = [(ff.butter_bandpass, {'lowcut': lowcut, 'highcut': highcut, 'fs': fs}),
             (ff.apply_notch_filter, {'notch_freq': notch_freq, 'fs': fs}),
             (ff.replace_outliers_with_threshold, {'num_std': num_std}),]

smeg_preprocess_data = smeg_data.copy()

for funct, params in tqdm(functions, desc="Procesando"):
  for i in range(smeg_dim[0]):
    for j in range(smeg_dim[1]):
      smeg_preprocess_data.iloc[i,j] = funct(smeg_data.iloc[i,j], **params)

smeg_preprocess_data
```

# Extracción de Características

```python
# Características que pueden extraerse
feature_funcs = {
    'std': ff.std,
    'rms': ff.rms,
    'iemg': ff.iemg,
    'mav': ff.mav,
    'wl': ff.wl,
    'log_detec': ff.log_detec,
    'ssi': ff.ssi,
    'fft': ff.fast_fourier_trans,
    'psd': ff.power_spect_dens,
    'mdf': ff.mdf,
    'zc': ff.zc,
}
```

```python
# Creamos un dataframe con columnas para cada característica y
# una columna adicional para los estímulos
features_data = pd.DataFrame(columns=list(feature_funcs.keys()) + ['stimulus'])

# Convertimos los datos del dataframe a datos tipos numpy
stimulus_np_data = stimulus_data.to_numpy()

for i in tqdm(range(smeg_dim[0]), desc="Extrayendo Características"):
  for j in range(smeg_dim[1]):
    features_row = {}
    series_data = smeg_preprocess_data.iloc[i, j].to_numpy()

    for name, func in feature_funcs.items():
      features_row[name] = func(series_data)

    features_row['stimulus'] = stimulus_data[i]
    features_data.loc[len(features_data)] = features_row

# Llenamos los datos NaN con 0, esto debido a la conversión de 0 a NaN
# en los dataframe
features_data.fillna(0, inplace=True)

features_data
```
